 GET USGS REAL TIME DATA

```
siteNum="413743089310102"
USGSparameter= "72019"
period="5"
url="http://waterservices.usgs.gov/nwis/iv/?format=json&sites="+siteNum+"&period=P"+period+"D&parameterCd="+USGSparameter+"&siteStatus=all"
print(url)
```

```
#open url
response = urllib.request.urlopen(url)
gw_data=json.loads(response.read())
```
